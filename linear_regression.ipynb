train-test-split

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,1]
y
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)

my_model=LinearRegression()
my_model.fit(x_train,y_train)
my_model.score(x_train,y_train)
```

```python
my_model.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,my_model.predict(x_train),color='blue')
plt.title('Salary vs experiance (Training set)')
plt.xlabel('years of experiance')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
y_prdict=my_model.predict(x_test)
plt.plot(x_test,y_prdict,color='blue')
plt.title('Salary vs experiance (Test set)')
plt.xlabel('years of experiance')
plt.ylabel('salary')
plt.show()
```

```python
c=[i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test,color='red',linestyle='-')
plt.plot(c,y_prdict,color='blue',linestyle='-')
plt.xlabel('salary')
plt.ylabel('experience')
plt.title('prediction')
plt.show()
```

```python
c=[i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test-y_prdict,color='green',linestyle='-')
plt.ylabel('error')
plt.xlabel('index')
plt.title('error value')
plt.show()
```

```python
mse=mean_squared_error(y_test,y_prdict)
rsq=r2_score(y_test,y_prdict)
print('mean square error=',mse)
print('r square=',rsq)
```

```python
print('intercept of the model :',my_model.intercept_)
print('coffecient of the model :',my_model.coef_)
```
